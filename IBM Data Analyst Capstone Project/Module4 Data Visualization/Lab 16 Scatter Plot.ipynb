#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
plt.figure(figsize=(12, 6))
plt.scatter(df['Age'], df['JobSatPoints_6'], alpha=0.5)
plt.xlabel('Age')
plt.xticks(rotation = 45, ha='right')
plt.ylabel('JobSat Points_6')
plt.title('A scatter plot of Age vs Job Sat Points 6')
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
plt.figure(figsize=(12, 6))
plt.scatter(df['ConvertedCompYearly'], df['JobSatPoints_6'], alpha=0.5)
plt.xlabel('ConvertedCompYearly')
plt.ylabel('JobSat Points_6')
plt.title('A scatter plot of Compensation vs Job Sat Points 6')
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
## Write your code here
!pip install seaborn
import seaborn as sns
```

```python
dfAS=df[['Age','JobSatPoints_6']]

agemap = {
'25-34 years old': 29,
'35-44 years old': 39,
'18-24 years old': 24,
'45-54 years old':49,
'55-64 years old':59,
'Under 18 years old':16,
'65 years or older': 70,
'Prefer not to say': None
}
dfAS['AgeNum'] = dfAS['Age'].map(agemap)
dfAS = dfAS[dfAS['AgeNum'].notnull()]
dfAS.dropna(inplace=True)


plt.figure(figsize=(12, 6))
sns.regplot(x='AgeNum', y='JobSatPoints_6', data=dfAS, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.xlabel('Age')
plt.ylabel('JobSat Points_6')
plt.title('A scatter plot of Age vs Job Sat Points 6')
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here
dfAW = df[['Age', 'YearsCodePro']]
dfAW['YearsCodeProNum'] = pd.to_numeric(dfAW['YearsCodePro'], errors='coerce')
plt.figure(figsize=(12,6))
sns.scatterplot(data=dfAW, x=dfAW['Age'], y=dfAW['YearsCodeProNum'])
plt.title('A scatter pklot of Age vs years Coded')
plt.xlabel('Age')
plt.ylabel('Years Coded profesionally')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
dfAS = df[['Age', 'JobSatPoints_6', 'ConvertedCompYearly']]
agemap = {
'25-34 years old': 29,
'35-44 years old': 39,
'18-24 years old': 24,
'45-54 years old':49,
'55-64 years old':59,
'Under 18 years old':16,
'65 years or older': 70,
'Prefer not to say': None
}
dfAS['AgeNum'] = dfAS['Age'].map(agemap)
dfAS = dfAS[dfAS['AgeNum'].notnull() & dfAS['ConvertedCompYearly'].notnull() & dfAS['JobSatPoints_6']]

plt.figure(figsize=(12, 6))
plt.scatter(dfAS['ConvertedCompYearly'], dfAS['JobSatPoints_6'],s=dfAS['AgeNum'].astype(float),alpha=0.5)
plt.xlabel('ConvertedCompYearly')
plt.xticks(rotation = 45, ha='right')
plt.ylabel('JobSat Points_6')
plt.title('A scatter plot of Age vs Job Sat Points 6')
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
dfLw = df[['LanguageHaveWorkedWith', 'JobSat']]
dfLw.dropna(subset='JobSat', inplace=True)
print(dfLw.isnull().sum())


#function to split the languages
def lang_split(languages):
    if pd.isna(languages):
        return []
    return str(languages).split(';')

#splitn the langueages
dfLw['LanguageWorkedWith'] = dfLw['LanguageHaveWorkedWith'].apply(lang_split)
have_langs = []
for sublist in dfLw['LanguageWorkedWith']:
    for lang in sublist:
        have_langs.append(lang)

#Create a DataFrame with individual languages and their corresponding job satisfaction
lang_data = []
for index, row in dfLw.iterrows():
    for lang in row['LanguageWorkedWith']:
        lang_data.append({'Language':lang,
                         'JobSat':row['JobSat']})
lang_df = pd.DataFrame(lang_data)

#aggregate the data
languagePopularity= lang_df['Language'].value_counts().reset_index()
languagePopularity.columns = ['Language', 'Popularity']

language_jobsat = lang_df.groupby('Language')['JobSat'].mean().reset_index()
language_jobsat.columns = ['Language', 'JobSat']

#merge the data
merged_data = pd.merge(languagePopularity, language_jobsat, on='Language')

#Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Language'], merged_data['JobSat'],s=merged_data['Popularity'], alpha=0.5)
plt.title('Scatter Plot of Programming Languages by Popularity and Job Satisfaction')
plt.xlabel('Language')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Average Job Satisfaction')
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
plt.figure(figsize=(12, 8))
for emp_type in df['Employment'].unique():
    subset = df[df['Employment'] == emp_type]
    plt.scatter(subset['ConvertedCompYearly'], subset['JobSatPoints_6'], 
                label=emp_type, alpha=0.7)
plt.title('Compensation vs. Job Satisfaction by Employment Type')
plt.xlabel('Yearly Compensation')
plt.ylabel('Job Satisfaction Score')
plt.legend()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 25, 35, 45, 100], 
                         labels=['18-25', '26-35', '36-45', '45+'])

plt.figure(figsize=(12, 8))
for country in df['Country'].unique():
    subset = df[df['Country'] == country]
    plt.scatter(subset['Age'], subset['YearsCodePro'], 
                label=country, alpha=0.7)
plt.title('Work Experience vs. Age by Country')
plt.xlabel('Age')
plt.ylabel('Years of Professional Code Experience')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
