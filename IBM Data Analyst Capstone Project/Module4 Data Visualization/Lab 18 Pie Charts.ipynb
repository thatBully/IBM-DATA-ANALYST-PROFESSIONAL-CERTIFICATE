### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
dfDB = df[['DatabaseWantToWorkWith']]

#split the database function
def split_databases(database):
    if pd.isna(database):
        return []
    return str(database).split(';')


dfDB['DatabaseWant'] = dfDB['DatabaseWantToWorkWith'].apply(split_databases)

want_db = []
for sublist in dfDB['DatabaseWant']:
    for db in sublist:
        want_db.append(db)
db_dataf = []
for index, row in dfDB.iterrows():
    for db in row['DatabaseWant']:
        db_dataf.append({'Individual Database':db})

db_df = pd.DataFrame(db_dataf)
db_df5 = db_df['Individual Database'].value_counts().head(5)
db_df5

plt.figure(figsize=(12,6))
plt.pie(db_df5,labels=db_df5.index, autopct='%1.1f%%')
plt.title('A pie chart showing the top 5 databases to work with')
plt.legend()
plt.show()

```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
top5DEV = df['DevType'].value_counts().head(5)
plt.figure(figsize=(12,12))
plt.pie(top5DEV, labels = top5DEV.index, autopct='%1.1f%%')
plt.title('A piechart showing the top 5 Developer types')
plt.legend(loc='best')
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here
topOs = df[['OpSysProfessional use']]

topOs['OpSysPro'] = topOs['OpSysProfessional use'].apply(split_databases)
topOs

Os =[]
for subList in topOs['OpSysPro']:
    for opSys in subList:
        Os.append(opSys)
db_OpSys =[]
for index, row in topOs.iterrows():
    for eachOs in row['OpSysPro']:
        db_OpSys.append({'Each Individual Os': eachOs})

db_OpSys = pd.DataFrame(db_OpSys)
top5 = db_OpSys['Each Individual Os'].value_counts().head(5)
plt.figure(figsize=(12, 12))
plt.pie(top5, labels=top5.index, autopct='%1.1f%%')
plt.title('A piechart showing the top 5 operation systems used professionally')
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here
topLang = df[['LanguageHaveWorkedWith']]

topLang['LanguageWorked'] = topLang['LanguageHaveWorkedWith'].apply(split_databases)

langHave =[]
for subList in topLang['LanguageWorked']:
    for lang in subList:
        langHave.append(lang)
db_Lang =[]
for index, row in topLang.iterrows():
    for eachLang in row['LanguageWorked']:
        db_Lang.append({'Individual Language': eachLang})

db_Lang = pd.DataFrame(db_Lang)
top5 = db_Lang['Individual Language'].value_counts().head(5)
plt.figure(figsize=(12, 12))
plt.pie(top5, labels=top5.index, autopct='%1.1f%%')
plt.title('A piechart showing the top 5 LAnguages')
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
##Write your code here
Colla = df[['NEWCollabToolsHaveWorkedWith']]
Colla['NewTools'] = Colla['NEWCollabToolsHaveWorkedWith'].apply(split_databases)

Tools = []
for sublist in Colla['NewTools']:
    for tool in sublist:
        Tools.append(tool)
ToolsData = []
for index, row in Colla.iterrows():
    for tool in row['NewTools']:
        ToolsData.append({'Individual Tools': tool})

ToolsDf = pd.DataFrame(ToolsData)
Tools5 = ToolsDf['Individual Tools'].value_counts().head(5)
print(Tools5)
plt.figure(figsize=(12,6))
plt.pie(Tools5, labels=Tools5.index ,autopct='%1.1f%%')
plt.title('A piechart showing the top 5 favorite collab tools to use')
plt.show()
```
